```python
def CountEq_LargestEq():
    x = Sig_Eqs['Country'].value_counts()#对各国出现的地震次数进行统计
    import numpy as np
    Sig_Eqs.sort_values(['Mag'],ascending=True)
    n=Sig_Eqs.groupby('Country')["Mag"].max()
    Sig_Eqs['Date']=Sig_Eqs['Year'].map(str)+"/"+Sig_Eqs['Mo'].map(str)+"/"+Sig_Eqs['Dy'].map(str)
    for i in n:
        for j in Sig_Eqs['Mag']:
            if i == j:
                m=Sig_Eqs[(Sig_Eqs.Mag==i)].index.tolist
                y=Sig_Eqs[Sig_Eqs[m,"Date"]]
        else:
            pass
        print(y)
    return(x,y)
CountEq_LargestEq()
```
